```python
import numpy as np
import pandas as pd 
```

```python
train = pd.read_csv('./titanic/train.csv')
test = pd.read_csv('./titanic/test.csv')

print('training data:', train.shape,'test data:', test.shape )
```

```python
full = train.append(test,ignore_index=True)

print(full.shape)
```

```python
full.head()
```

```python
full.describe()
```

```python
full.info()
```

```python
full['Fare'] = full['Fare'].fillna(full['Fare'].mean())
full['Age'] = full['Age'].fillna(full['Age'].mean())
```

```python
full.groupby('Embarked').size()
```

```python
full['Embarked'].mode()
```

```python
full['Embarked'] = full['Embarked'].fillna('S')
```

```python
full['Cabin'] = full['Cabin'].fillna('U')
```

```python
full['Embarked'].value_counts()
```

```python
full.info()
```

```python
sex_mapDict={'male':1,
            'female':0}

full['Sex'] = full['Sex'].map(sex_mapDict)
full.head()
```

```python
full['Embarked'].head()
```

```python
embarkedDf = pd.DataFrame()

embarkedDf = pd.get_dummies(full['Embarked'], prefix='Embarked')
embarkedDf.head()
```

```python
full= pd.concat([full,embarkedDf],axis=1)
full.drop('Embarked', axis = 1, inplace = True)
full.head()
```

```python
full['Name'].head(10)
```

```python
def gettitle(name):
    str1 = name.split(',')[1]
    str2 = str1.split('.')[0]
    str3 = str2.strip()
    return str3
```

```python
TitleDf = pd.DataFrame()
TitleDf = full['Name'].map(gettitle)
TitleDf.head()
```

```python
TitleDf.value_counts()
```

```python
title_mapDict={"Mr":"Mr","Miss":"Miss","Mrs":"Mrs","Master":"Master","Rev":"Officer","Dr":"Officer","Col":"Officer","Major":"Officer",
              "Mlle":"Miss","Ms":"Mrs","Dona":"Royalty","Lady":"Royalty","Jonkheer":"Royalty","Capt":"Officer","Sir":"Royalty",
              "Mme":"Mrs","the Countess":"Royalty","Don":"Royalty"}
```

```python
TitleDf = TitleDf.map(title_mapDict)
TitleDf = pd.get_dummies(TitleDf)
TitleDf.head()
```

```python
full = pd.concat([full, TitleDf], axis=1)
full.drop('Name', axis=1, inplace=True)
```

```python
full['Cabin'] = full['Cabin'].map(lambda c:c[0])
CabinDf = pd.DataFrame()
CabinDf = pd.get_dummies(full['Cabin'], prefix='Cabin')
full= pd.concat([full, CabinDf], axis=1)
full.drop('Cabin', axis=1, inplace=True)
```

```python
familyDf = pd.DataFrame()
familyDf['familysize'] = full['Parch'] + full['SibSp'] + 1
familyDf['family_single']=familyDf['familysize'].map(lambda s:1 if s==1 else 0)
familyDf['family_small']=familyDf['familysize'].map(lambda s:1 if 2<=s<=4 else 0)
familyDf['family_large']=familyDf['familysize'].map(lambda s:1 if s>5 else 0)
```

```python
PclassDf = pd.DataFrame()
PclassDf = pd.get_dummies(full['Pclass'] , prefix='Pclass' )
PclassDf.head()
full = pd.concat([full,PclassDf],axis=1)
full.drop('Pclass',axis=1,inplace=True)
full.head()
```

```python
full = pd. concat([full, familyDf], axis=1)
full.shape
```
